```python
import pandas as pd
import numpy as np
```

# 数据清洗

```python
import glob
import os

# 搜索匹配的文件
fpath= glob.glob("./数据保存位置/价格表*.xlsx")[0]
pdata=pd.read_excel(fpath)
pdata
```

```python
#判断名字有无缺失
print(np.all(pd.notnull(pdata['品名'])))
#判断平均值有无缺失,如果有则补充
if np.any(pd.isnull(pdata['平均价'])):
    #计算最高值和最低值的平均值
    fill_value=(pdata['最低价']+pdata['最高价'])/2
    #补充平均值
    pdata['平均价'].fillna(fill_value, inplace=True)
```

```python
#判断数据类型
pdata.dtypes
```

```python
#修改数据类型
#修改日期的数据类型
pdata['发布日期'] = pd.to_datetime(pdata['发布日期'])
#检查是否修改成功
print(pdata['发布日期'].dtypes)
```

```python
# 删除重复行
pdata_unique = pdata.drop_duplicates()
```

```python
#将产地每种蔬菜的产地分开
# 定义关键字
keywords=['京','津','沪','渝','冀','晋','辽','吉','黑','苏','浙','皖','闽','赣','鲁','豫','鄂','湘','粤','桂','琼','川','黔','滇','藏','陕','甘','青','宁','新','港','澳','台']
new_keywords=[]
for i in keywords:
    count=pdata['产地'].str.count(i).sum()
    print(i,'出现:',count)
    #将没出现的产地删除
    if count > 0:
        new_keywords.append(i)
print(new_keywords)
with pd.ExcelWriter('./数据保存位置/价格表-2025-06-27-00-32-40价格表.xlsx', engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    for keyword in keywords:
        # 筛选出包含关键字的行
        filtered_pdata = pdata[pdata['产地'].str.contains(keyword, na=False)]
        # 如果筛选结果不为空，则将其保存到对应的工作表中
        if not filtered_pdata.empty:
            filtered_pdata.to_excel(writer, sheet_name=keyword, index=False)

print(f"数据已保存")
```

# 数据可视化

```python
import matplotlib.pyplot as plt
import seaborn as sns
# 设置支持中文的字体
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
```

```python
pdata.describe
```

```python
# 计算四分位数
Q1 = pdata['平均价'].quantile(0.25)
Q3 = pdata['平均价'].quantile(0.75)
IQR = Q3 - Q1
# 定义数值范围
low = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
# 限制数据范围
filtered_data = pdata[(pdata['平均价'] >= low) & (pdata['平均价'] <= upper)]
# 箱线图
plt.figure(figsize=(10, 6))
plt.boxplot(filtered_data['平均价'], vert=False)
plt.title('市场蔬菜价格分布')
plt.xlabel('平均价')
plt.show()
```

```python
#分析前5蔬菜省份
# 读取所有工作表的数据
xls = pd.ExcelFile(fpath)
# 排除第一张总表
sheet_names = xls.sheet_names[1:]
dataframes = {sheet: pd.read_excel(fpath, sheet_name=sheet) for sheet in sheet_names}
#统计各省份蔬菜数量
sheet_counts = {sheet: df['品名'].count() for sheet, df in dataframes.items()}
# 转换为DataFrame
count_df = pd.DataFrame(list(sheet_counts.items()), columns=['省份', '蔬菜数量'])
# 取前五省份
top_5 = count_df.sort_values(by='蔬菜数量', ascending=False).head(5)
#绘制条形图
plt.figure(figsize=(12, 8))
bars = plt.bar(top_5['省份'], top_5['蔬菜数量'], color='skyblue')
plt.title('市场进口蔬菜前5大省', fontsize=16)
plt.xlabel('省份', fontsize=12)
plt.ylabel('蔬菜数量', fontsize=12)
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 1, 
             f'{height}', ha='center', va='bottom', fontsize=10)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
xls = pd.ExcelFile(fpath)

vegetable_counts = []  # 用于存储每个省份的蔬菜种类数量
province_names = []  # 用于存储省份名称

for sheet_name in xls.sheet_names:
    if sheet_name != 'Sheet1':  # 跳过总表 Sheet1
        df = pd.read_excel(xls, sheet_name=sheet_name)
        
        # 统计蔬菜种类数量（即表的行数）
        vegetable_count = len(df)
        
        # 将结果存储到列表中
        vegetable_counts.append(vegetable_count)
        province_names.append(sheet_name)

# 绘制饼图
plt.figure(figsize=(8, 8))  # 设置图形大小
plt.pie(vegetable_counts, labels=province_names, autopct='%1.1f%%', startangle=140)
plt.title("各省份进口蔬菜种类比例")
plt.show()
```

```python
df = pd.read_excel(fpath)
# 筛选品名为大白菜且日期在 2025-06-13 到 2025-06-19 之间的数据
start_date = '2025-06-13'
end_date = '2025-06-19'
filtered_df = df[(df['品名'] == '大白菜') & (df['发布日期'] >= start_date) & (df['发布日期'] <= end_date)]

# 按日期排序
filtered_df = filtered_df.sort_values(by='发布日期')

# 绘制价格变化趋势图
plt.figure(figsize=(10, 6))  # 设置图形大小
plt.plot(filtered_df['发布日期'], filtered_df['平均价'], marker='o', linestyle='-', label='平均价')
plt.plot(filtered_df['发布日期'], filtered_df['最高价'], marker='^', linestyle='--', label='最高价')
plt.plot(filtered_df['发布日期'], filtered_df['最低价'], marker='v', linestyle='--', label='最低价')
plt.title('大白菜一周价格变化趋势')
plt.xlabel('发布日期')
plt.ylabel('价格')
plt.grid(True)  # 添加网格线
plt.xticks(rotation=45)  # 旋转日期标签，避免重叠
plt.legend()  # 添加图例
plt.tight_layout()  # 自动调整子图参数，以确保子图之间有足够空间
plt.show()
```

```python
df = pd.read_excel(fpath, sheet_name='Sheet1')
df['发布日期'] = pd.to_datetime(df['发布日期'], errors='coerce')
def classify_season(month):
    if month in [12, 1, 2]:
        return '冬季'
    elif month in [3, 4, 5]:
        return '春季'
    elif month in [6, 7, 8]:
        return '夏季'
    else:
        return '秋季'

#分类
df['季节'] = df['发布日期'].dt.month.apply(classify_season)
# 选出大白菜
filtered_df = df[df['品名'] == '大白菜']
# 计算每个季节的平均价格
seasonal_avg_prices = filtered_df.groupby('季节')['平均价'].mean().reset_index()

# 散点图
plt.figure(figsize=(10, 6))
for i, row in seasonal_avg_prices.iterrows():
    plt.scatter(row['季节'], row['平均价'], s=100, label=row['季节'])

plt.title('大白菜不同季节的平均价格')
plt.xlabel('季节')
plt.ylabel('平均价')
plt.grid(True)
plt.legend(title='季节')
plt.tight_layout()
plt.show()
```
